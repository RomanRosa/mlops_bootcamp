- Divorced: 84 (49%)
- Married: 86 (51%)

**Data Collection**:  
Participants completed the Personal Information Form and Divorce Predictors Scale.

**Analysis Methods**:  
The efficacy of the Divorce Predictors Scale (DPS) was investigated using:
- Multilayer Perceptron Neural Network
- C4.5 Decision Tree Algorithm

#### Visualization ###

```python
# Countplot with different colors for each class
sns.countplot(x='Class', data=df, palette='Set1')

# Enhancing the plot
plt.title('Count of Different Classes')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
# Correlation values 
corr = df.corr()['Class'].drop('Class')
sort_corr=corr.abs().sort_values(ascending=False)[:20]
sort_corr
```

```python
# Show plot to compare correlation values
sns.barplot(x=sort_corr.index, y=sort_corr, palette='viridis')
plt.xticks(rotation=90)
plt.xlabel('Features')
plt.ylabel('Absolute Correlation')
plt.show()
```
